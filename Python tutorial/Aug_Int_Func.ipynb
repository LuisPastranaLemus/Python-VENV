### __Python augmented assignment Operators and Integrated functions__

```python
items = 10
items *= 5
print(items)

items = 10
items /= 5
print(items)

items = 10
items -= 5
print(items)

items = 10
items += 5
print(items)

items = 10
items %= 5
print(items)

items = 10
items **= 5
print(items)

items = 10
items //= 5
print(items)

items = 10
items &= 5
print(items)

items = 10
items |= 5
print(items)

items = 10
items ^= 5
print(items)

items = 10
items >>= 5
print(items)

items = 10
items <<= 5
print(items)
```

```python
movies_duration = [142, 175, 152, 195, 201, 154, 178, 139, 133, 126]

total_duration = sum(movies_duration)
print("Sum: ", total_duration)

max_duration = max(movies_duration)
print("Max", max_duration)

min_duration = min(movies_duration)
print("Min", min_duration)

total_movies = len(movies_duration)
print("Len: ", total_movies)

print("Abs: ", abs(-5))

print("Round: ", round(3.14159265359, 2))

print("Sorted: ", sorted([9,4,7,2,1,5,3,6,8]))
print("Sorted reverse", sorted([9,4,7,2,1,5,3,6,8], reverse=True))
print("Reversed: ", list(reversed([1, 2, 3])))

print("All: ", all([True, True, False])) # Returns True if all elements of the iterable are true (or if the iterable is empty).

print("Any: ", any([True, True, False])) # Returns True if any element of the iterable is true.

print("Enumerate: ", list(enumerate(['a', 'b', 'c']))) # Returns a list of tuples containing indices and values from the iterable.

print("Zip: ", zip([1, 2, 3], ['a', 'b', 'c'])) # Combines multiple iterables (e.g., lists) element-wise, returning tuples.

print("Is instance: ", isinstance(5, int)) # Returns True if the specified object is of the specified type, otherwise False.

print(chr(65))  # Returns a character from the specified Unicode code.


numbers = [1, 2, 3, 4, 5, 6]

# Define a function that checks if a number is odd
def is_odd(num):
    return num % 2 != 0

# Use filter to get only the odd numbers
odd_numbers = filter(is_odd, numbers)

# Convert filter object to a list and print the result
print("filter:", list(odd_numbers))



# List of temperatures in Celsius
celsius_temps = [0, 20, 30, 40, 100]

# Define a function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map to apply the conversion function to each temperature
fahrenheit_temps = map(celsius_to_fahrenheit, celsius_temps) # The map() function applies this conversion to each item in the list celsius_temps.

# Convert the map object to a list and print the result
print("Map: ", list(fahrenheit_temps))
```
